```python
im = np.array(pil_img)
nbr_bins = 256
```

```python
imhist, bins = np.histogram(im.flatten(), nbr_bins)

cdf = imhist.cumsum()
cdf = 255 * cdf / cdf[-1]
```

```python
im2 = np.interp(im.flatten(), bins[:-1], cdf)
```
